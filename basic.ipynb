```python
# importar pacotes
import numpy as np
import matplotlib.pyplot as plt
```

```python
# gerar valores demonstrativos
np.random.seed(42)
x = np.arange(10)
y = np.random.normal(size=10)

# plotar os valores
plt.plot(x, y)

# para definir o título do plot, utilize plot.title('Nome do Plot')
plt.title("Exemplo de Título")

# também é possível nomear os eixos, utilizando xlabel e ylabel
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

# mostando o plot
plt.show()
```

```python
# criando o plot
plt.plot(x, x, label="Linear")

# adicionando ao plot
plt.plot(x, x**2, label="Quadrado")

# definindo títulos
plt.title("Abordagem Pyplot")
plt.ylabel("Valores de y")
plt.xlabel("Valores de x")

# plotando a legenda
plt.legend()

# mostrando o plot
plt.show()
```
